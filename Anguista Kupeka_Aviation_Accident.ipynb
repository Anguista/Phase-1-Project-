# 3. Data Cleaning

```python
#First we drop all the duplicates from the data set.
df.drop_duplicates()
df
```

```python
#Then we drop all the unnecessary columns.
df.drop(columns=['Event.Id', 'Investigation.Type', 'Accident.Number', 'Location', 'Country', 
                                 'Latitude', 'Longitude', 'Airport.Code', 'Airport.Name', 'Registration.Number', 
                                 'Amateur.Built', 'FAR.Description', 'Schedule', 'Air.carrier', 'Report.Status', 
                                 'Publication.Date'], errors='ignore', inplace=True)
df
```

```python

#First we will print the columns to see the names of all the columns we have left after dropping all the unnecessary columns.
print(df.columns)
```

```python
#Next we make sure that all the columns are spelled correctly.
df.rename(columns={ 'Event.Date' : 'Event Date', 'Injury.Severity' : 'Injury Severity', 'Aircraft.damage' : 'Aircraft Damage', 'Aircraft.Category' : 'Aircraft Category', 'Make' : 'Aircraft Make', 'Model' : 'Aircraft Model', 'Number.of.Engines' : 'Number of Engines', 'Engine.Type' : 'Engine Type', 'FAR.Description' : 'FAR Description', 'Purpose.of.flight' : 'Purpose of flight', 'Total.Fatal.Injuries' :'Total Fatal Injuries','Total.Serious.Injuries' : 'Total Serious Injuries', 'Total.Minor.Injuries' :'Total Minor Injuries', 'Total.Uninjured' : 'Total Un-injured', 'Weather.Condition' : 'Weather Condition', 'Broad.phase.of.flight' :'Broad Phase of Flight'}, inplace=True)
df
```

```python
#Afterwards, we go ahead and check if we have any null or missing values
# then sum them so as to decide how to manipulate them
df.isnull().values.any()
df.isnull().sum()
```

```python
#In order to fill the null or missing values appropriately we check the all the column data type
#First we check for all the numerical columns 
numerical_columns = df.select_dtypes(include=['number']).columns
print(numerical_columns)

```

```python
mode = df['Number of Engines'].mode()

print("Number of Engines:", mode.tolist())
```

```python
# Calculate the mean
mean = df['Number of Engines'].mean()

print("Mean of the Number of Engines':", mean)
```

```python
#We can see that the following columns: 'Total Minor Injuries', 'Total Un-injured', 'Number of Engines', 'Total Fatal Injuries', 'Total Serious Injuries', are numerical columns
#We proceed to fill the null values with a constant number 0 so as not to disrupt the integrity of the data when we analyse.
df['Number of Engines'] = df['Number of Engines'].fillna(0)
df['Total Fatal Injuries'] = df['Total Fatal Injuries'].fillna(0)
df['Total Serious Injuries'] = df['Total Serious Injuries'].fillna(0)
df['Total Minor Injuries'] = df['Total Minor Injuries'].fillna(0)
df['Total Un-injured'] = df['Total Un-injured'].fillna(0)
df

```

```python
#Then we check for all the categorical columns.
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)

```

```python
#We can see that the following columns: 'Event Date', 'Injury Severity', 'Aircraft Damage', 'Aircraft Category','Aircraft Make', 'Aircraft Model', 'Engine Type', 'Purpose of flight','Weather Condition', 'Broad Phase of Flight' are categorical columns
#We proceed to fill the null values with a constant value 'Not identified' so as not to disrupt the integrity of the data when we analyse.
for col in categorical_columns:
    df[col] = df[col].fillna('Not identified')
df
```

```python
#Next we check for inconsistencies in the casing in the different categorical columns
for col in categorical_columns:
    print(f"Unique values in {col}: {df[col].unique()}")
```

```python
for col in categorical_columns:
    print(f"Unique values in {col}: {df[col].unique()}")
df
```

```python
#Then we convert all the values in all the categorical columns to lower case for consistency.
for col in categorical_columns:
    df[col] = df[col].str.lower()
    df
```

```python
#Next we capitalize all the first letters of every word of all the values in all the categorical columns.
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_columns:
    df[col] = df[col].str.title()
    df
```

```python
#Next we correct the spelling of some values in a specific column like column 'Aircraft Make'
df.loc[df['Aircraft Make'].str.contains('Cesna', case=False, na=False), 'Aircraft Make'] = 'Cessna'
df.loc[df['Aircraft Category'].str.contains('Wsft', case=False, na=False), 'Aircraft Category'] = 'Weight-Shift'
df.loc[df['Weather Condition'].str.contains('Vmc', case=False, na=False), 'Weather Condition'] = 'VMC'
df.loc[df['Weather Condition'].str.contains('Unk', case=False, na=False), 'Weather Condition'] = 'UNK'
df.loc[df['Weather Condition'].str.contains('Imc', case=False, na=False), 'Weather Condition'] = 'IMC'


```

```python
#Next we check for odd symbols in all the categorical columns.

for col in categorical_columns:
    df_with_symbols = df[df[col].str.contains(r'[^a-zA-Z0-9\s]', regex=True)]
    if not df_with_symbols.empty:
        print(f"Odd symbols found in column: {col}")
        print(df_with_symbols)

```

```python
#Next we remove all the odd symbols in the categorical columns.
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

for col in categorical_columns:
    df[col] = df[col].str.replace(r'[#$@]', '', regex=True)
```

```python
#Next we remove all the leading and trailing white spaces in multiple columns.
for col in categorical_columns:
    df[col] = df[col].str.strip()
```

```python
#For ease in analysis, and creating visualizations to give efficient insights on the efficiency and safety of the aircraft we are going to merge the following columns together and rename them: 'Injury Severity' and 'Aircraft Damage' into 'Accident Severity', 'Aircraft Category' and 'Aircraft Make' and 'Aircraft Model'  into 'Aircraft Type', 'Number of Engines' and 'Engine Type' into 'Engine Configuration', 'Purpose of flight' and 'Broad Phase of Flight' into 'Flight Operation Context', 'Weather Condition' and 'Event Date' into 'Weather at Event'.

#First we will ensure that all columns are strings before concatenating

df['Number of Engines'] = df['Number of Engines'].astype(str)
df['Engine Type'] = df['Engine Type'].astype(str)
df['Injury Severity'] = df['Injury Severity'].astype(str)
df['Aircraft Damage'] = df['Aircraft Damage'].astype(str)
df['Aircraft Category'] = df['Aircraft Category'].astype(str)
df['Aircraft Make'] = df['Aircraft Make'].astype(str)
df['Aircraft Model'] = df['Aircraft Model'].astype(str)
df['Purpose of flight'] = df['Purpose of flight'].astype(str)
df['Broad Phase of Flight'] = df['Broad Phase of Flight'].astype(str)
df['Weather Condition'] = df['Weather Condition'].astype(str)
df['Event Date'] = df['Event Date'].astype(str)

```

```python
#Then we will proceed to merge the multiiple columns.
df['Engine Configuration'] = df['Number of Engines'] + '-' + df['Engine Type']
df['Accident Severity'] = df['Injury Severity'] + '-' + df['Aircraft Damage']
df['Aircraft Type'] = df['Aircraft Category'] + '-' + df['Aircraft Make'] + '-' + df['Aircraft Model']
df['Flight Operation Context'] = df['Purpose of flight'] + '-' + df['Broad Phase of Flight']
df['Weather at Event'] = df['Weather Condition'] + '-' + df['Event Date']


print(df)
```
